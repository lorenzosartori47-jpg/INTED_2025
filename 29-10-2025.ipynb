```python
def obter_dia_da_semana_condicional():
    """Pede um nÃºmero de 1 a 7 e usa condiÃ§Ãµes para retornar o dia."""

    print("--- Tradutor de NÃºmero para Dia da Semana ---")

    try:
        # Pede o nÃºmero ao usuÃ¡rio
        numero = int(input("Digite um nÃºmero de 1 a 7: "))

        print(f"\nO nÃºmero {numero} corresponde Ã : ")

        if numero == 1:
            dia = "Domingo"
        elif numero == 2:
            dia = "Segunda-feira"
        elif numero == 3:
            dia = "TerÃ§a-feira"
        elif numero == 4:
            dia = "Quarta-feira"
        elif numero == 5:
            dia = "Quinta-feira"
        elif numero == 6:
            dia = "Sexta-feira"
        elif numero == 7:
            dia = "SÃ¡bado"
        else:
            dia = "âŒ Erro: NÃºmero fora do intervalo (1 a 7)."

        print(f"**{dia}**.")

    except ValueError:
        print("\nâŒ Erro: Entrada invÃ¡lida. Por favor, digite um nÃºmero inteiro.")

# Chama a funÃ§Ã£o para executar
obter_dia_da_semana_condicional()
```

```python
def verificar_ano_bissexto_detalhado():
    """Pede um ano e verifica se Ã© bissexto usando uma estrutura if/elif/else."""

    print("--- Verificador de Ano Bissexto ---")

    while True:
        try:
            # Pede a entrada do usuÃ¡rio
            ano = int(input("Digite um ano (ex: 2024): "))
            break
        except ValueError:
            print("ðŸš¨ Erro: Entrada invÃ¡lida. Por favor, digite um nÃºmero inteiro.")

    if ano % 400 == 0:
        # 1. Regra de exceÃ§Ã£o principal: DivisÃ­vel por 400 (ex: 2000)
        e_bissexto = True
    elif ano % 100 == 0:
        # 2. Regra de exceÃ§Ã£o: DivisÃ­vel por 100, mas nÃ£o por 400 (ex: 1900)
        e_bissexto = False
    elif ano % 4 == 0:
        # 3. Regra padrÃ£o: DivisÃ­vel por 4 (ex: 2024)
        e_bissexto = True
    else:
        # 4. Outros anos (ex: 2023)
        e_bissexto = False

    if e_bissexto:
        print(f"\nâœ… O ano {ano} **Ã©** bissexto.")
    else:
        print(f"\nâŒ O ano {ano} **nÃ£o Ã©** bissexto.")

# Executa a funÃ§Ã£o
verificar_ano_bissexto_detalhado()
```

```python
def calcular_salario_com_extras():
    """
    Calcula o salÃ¡rio total com adicional de 50% para horas extras (> 40h).
    Pronto para rodar no Google Colab.
    """

    print("--- Calculadora de SalÃ¡rio com Horas Extras ---")
    LIMITE_HORAS = 40.0
    TAXA_EXTRA = 1.5  # 50% de adicional (1.0 da hora normal + 0.5 de adicional)

    # Coleta e ValidaÃ§Ã£o das Horas Trabalhadas
    while True:
        try:
            horas_trabalhadas = float(input("Digite o total de horas trabalhadas na semana: "))
            if horas_trabalhadas >= 0:
                break
            else:
                print("ðŸš¨ Erro: O nÃºmero de horas nÃ£o pode ser negativo.")
        except ValueError:
            print("ðŸš¨ Erro: Entrada invÃ¡lida. Por favor, digite um nÃºmero (ex: 45.5).")

    # Coleta e ValidaÃ§Ã£o do Valor por Hora
    while True:
        try:
            valor_por_hora = float(input("Digite o valor da hora de trabalho (R$): "))
            if valor_por_hora >= 0:
                break
            else:
                print("ðŸš¨ Erro: O valor por hora nÃ£o pode ser negativo.")
        except ValueError:
            print("ðŸš¨ Erro: Entrada invÃ¡lida. Por favor, digite um nÃºmero (ex: 30.50).")

    # --- LÃ³gica de CÃ¡lculo ---

    if horas_trabalhadas <= LIMITE_HORAS:
        # Caso 1: Sem horas extras
        salario_total = horas_trabalhadas * valor_por_hora
        print(f"\nâœ… SalÃ¡rio Total (Sem Extras): R$ **{salario_total:.2f}**")

    else:
        # Caso 2: Com horas extras

        # SalÃ¡rio base pelas 40 horas
        salario_base = LIMITE_HORAS * valor_por_hora

        # CÃ¡lculo das horas e valor extras
        horas_extras = horas_trabalhadas - LIMITE_HORAS
        valor_horas_extras = horas_extras * valor_por_hora * TAXA_EXTRA

        salario_total = salario_base + valor_horas_extras

        # ExibiÃ§Ã£o Detalhada
        print("\n--- DETALHES DO CÃLCULO ---")
        print(f"âœ”ï¸ SalÃ¡rio Base ({LIMITE_HORAS:.0f}h): R$ {salario_base:.2f}")
        print(f"âž• Horas Extras ({horas_extras:.2f}h): R$ {valor_horas_extras:.2f} (50% Adicional)")
        print("-" * 35)
        print(f"ðŸ’° **SALÃRIO TOTAL A PAGAR:** R$ **{salario_total:.2f}**")

# Executa a funÃ§Ã£o
calcular_salario_com_extras()
```
