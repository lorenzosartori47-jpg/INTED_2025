```python
def verificar_ano_bissexto_detalhado():
    """Pede um ano e verifica se √© bissexto usando uma estrutura if/elif/else."""

    print("--- Verificador de Ano Bissexto ---")

    while True:
        try:
            # Pede a entrada do usu√°rio
            ano = int(input("Digite um ano (ex: 2024): "))
            break
        except ValueError:
            print("üö® Erro: Entrada inv√°lida. Por favor, digite um n√∫mero inteiro.")

    if ano % 400 == 0:
        # 1. Regra de exce√ß√£o principal: Divis√≠vel por 400 (ex: 2000)
        e_bissexto = True
    elif ano % 100 == 0:
        # 2. Regra de exce√ß√£o: Divis√≠vel por 100, mas n√£o por 400 (ex: 1900)
        e_bissexto = False
    elif ano % 4 == 0:
        # 3. Regra padr√£o: Divis√≠vel por 4 (ex: 2024)
        e_bissexto = True
    else:
        # 4. Outros anos (ex: 2023)
        e_bissexto = False

    if e_bissexto:
        print(f"\n‚úÖ O ano {ano} **√©** bissexto.")
    else:
        print(f"\n‚ùå O ano {ano} **n√£o √©** bissexto.")

# Executa a fun√ß√£o
verificar_ano_bissexto_detalhado()
```
